```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('afp_links.csv')
```

```python
df
```

```python
afp_links_df = df.url.tolist()
afp_links_df
```

## Get the headline

```python
for url in afp_links_df:
    response = requests.get(url)
    doc = BeautifulSoup(response.content, "html.parser")
    title = doc.find(class_='content-title')
    print(title.text)
```

## Get publish date 

```python
for url in afp_links_df:
    response = requests.get(url)
    doc = doc = BeautifulSoup(response.content, "html.parser")
    datetime = doc.find(class_='date')
    print(datetime.text)
```

## Get fact-check details

```python
for url in afp_links_df:
    response = requests.get(url)
    doc = BeautifulSoup(response.content, "html.parser")
    details = doc.select_one('h3')
    if details:
        print(details.string)
        print("-------------------------")
    else:
        print("None")
```

## Add everything to a list

```python
afp_list = []
for url in tqdm(afp_links_df):
    afp_dict = {}
    response = requests.get(url)
    doc = BeautifulSoup(response.content, "html.parser")
    
    title = doc.find(class_='content-title')
    afp_dict['headline'] = title.text
    
    date = doc.find(class_='date')
    afp_dict['datetime'] = date.text
    
    details = doc.select_one('h3')
    if details:
        afp_dict['fact_check_details'] = details.string
    else:
        afp_dict['fact_check_details'] = "NA"
    
    afp_list.append(afp_dict)  
afp_list
```

```python
len(afp_list)
```

## Make a dataframe of the list

```python
df_new = pd.DataFrame(afp_list)
df_new.head()
```

## Merge df and df_new

```python
afp_df = pd.concat([df, df_new], axis=1, join='inner')
afp_df
```
