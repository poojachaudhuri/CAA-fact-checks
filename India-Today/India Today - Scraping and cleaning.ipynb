```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

df = pd.read_csv('india_today.csv')
df.head()
```

```python
df.shape
```

```python
india_today_urls = df.url.tolist()
india_today_urls
```

```python
india_today_list = []

for link in india_today_urls:
    india_today_dict = {}
    response = requests.get(link)
    doc = BeautifulSoup(response.content, "html.parser")
    
    # get titles
    title = doc.find("h1", class_="jsx-99cc083358cc2e2d Story_strytitle__MYXmR")
    
    # get date 
    date = doc.find(class_='jsx-99cc083358cc2e2d strydate')
    
    # get fact-check conclusion
    details = doc.find(class_='factcheckDetail_conclusion__content__gFnx5')
    
    # put them all in the dictionary 
    india_today_dict['headline'] = title.string
    india_today_dict['datetime'] = date.text
    india_today_dict['fact_check_details'] = details.text
    
    # append the dict to the list
    india_today_list.append(india_today_dict)
    
india_today_list
```

```python
df2 = pd.DataFrame(india_today_list)
df2.head()
```

```python
df.head()
```

```python
india_today_df = pd.concat([df, df2], axis=1, join='inner')
india_today_df
```

```python
india_today_df.head()
```

```python
india_today_df['organization'] = "India Today"
india_today_df
```

```python
india_today_df.to_csv('india_today_df.csv', index=False)
```
