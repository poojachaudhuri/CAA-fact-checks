## Make a function

```python
from playwright.async_api import async_playwright
playwright = await async_playwright().start()
browser = await playwright.chromium.launch(headless = False)
import time
# import tqdm
import pandas as pd
async def list_of_links(url):
    page = await browser.new_page()
#     link = url
    await page.goto(url)    
    link_list = []
    while True:
        try:
            results = page.locator('h2.entry-title.e-t a')
            time.sleep(1)
            count = await results.count()
            for i in range(count):
                links = results.nth(i)
                url = await links.get_attribute('href')
                link_list.append(url)

            async with page.expect_navigation():
                await page.locator('.ti__arrow-left').click(timeout=5000)
            print("Pressing the button")
            print("---------")
        except:
            break
    
    
    df_links = pd.DataFrame(link_list)
    return df_links
```

### Scrape the website

```python
url = "https://www.altnews.in/no-the-elderly-man-pelting-stone-in-kanpur-is-not-muslim-cleric-from-muzaffarnagar/"
response = requests.get(url)
doc = BeautifulSoup(response.content, "html.parser")
```

```python
# get the title for the first article 

doc.find(class_='headline entry-title e-t').text
```

```python
# get the date of publish for the first article 

doc.find(class_='entry-date published').text
```

```python
# do it for all urls and put it in a list 

alt_list = []
for url in tqdm(altnews_list):
    alt_dict = {}
    response = requests.get(url)
    doc = BeautifulSoup(response.content, "html.parser")
    
    title = doc.find(class_='headline entry-title e-t')
    alt_dict['headline'] = title.text
    
    date = doc.find(class_='entry-date published')
    alt_dict['datetime'] = date.text
    
    alt_list.append(alt_dict)
    
alt_list  
```

```python
altnews_all_links = pd.DataFrame(alt_list)
altnews_all_links
```
