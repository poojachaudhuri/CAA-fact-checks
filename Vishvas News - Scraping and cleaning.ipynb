## Put everything in a list of dictionary

```python
vishvas_news_list = []
for url in vishvas_news_urls:
    vishvas_news_dict = {}
    response = requests.get(url)
    doc = BeautifulSoup(response.content, "html.parser")
    
    # get titles
    title = doc.find("h1", class_="article-heading")
    vishvas_news_dict['headline'] = title.string
    
    # get date
    date = doc.find(class_="updated")
    vishvas_news_dict['datetime'] = date.text
    
    
    # get fact-check details
    details = doc.select_one('.claim-review li')
    if details:
        vishvas_news_dict['fact_check_details'] = details.text
    else:
        vishvas_news_dict['fact_check_details'] = "NA"
    
    vishvas_news_list.append(vishvas_news_dict)
```

```python
vishvas_news_list
```

```python
df2 = pd.DataFrame(vishvas_news_list)
df2.head()
```

## Merge both df

```python
vishvas_news_df = pd.concat([df, df2], axis=1, join='inner')
vishvas_news_df
```

```python
vishvas_news_df['organization'] = "Vishvas News"
vishvas_news_df
```

## Clean the data

```python
vishvas_news_df['headline'] = vishvas_news_df['headline'].apply(lambda x: x.replace('\n', ''))
vishvas_news_df
```

```python
vishvas_news_df['datetime'] = vishvas_news_df['datetime'].apply(lambda x: x.replace('\n', ''))
vishvas_news_df.sample(20)
```
