```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import re
```

```python
df = pd.read_csv('quint_links.csv')
df
```

```python
df.columns = ['url']
df
```

```python
df_list = df.url.tolist()
df_list
```

```python
url = 'https://www.thequint.com/news/webqoof/nirbhaya-helpline-number-launched-fact-check#read-more'
response = requests.get(url)
doc = BeautifulSoup(response.content, "html.parser")
```

```python
quint_list = []
for url in tqdm(df_list):
    quint_dict = {}
    response = requests.get(url)
    doc = BeautifulSoup(response.content, "html.parser")
    
    title = doc.find('h1')
    quint_dict['headline'] = title.string
    
    date = doc.find(class_='_1yYGD')
    quint_dict['datetime'] = date.string
    
    details = doc.find('h2')
    quint_dict['fact_check_details'] = details.string
    
    quint_list.append(quint_dict)
quint_list
```

```python
len(quint_list)
```

```python
df2 = pd.DataFrame(quint_list)
df2
```

```python
df2 = df2.rename(columns={'fact_checK_details': 'fact_check_details'})
```

```python
quint_df = pd.concat([df, df2], axis=1, join='inner')
quint_df
```

```python
quint_df['organization'] = "The Quint"
quint_df
```

```python
quint_df.datetime
```

```python
quint_df.datetime.str.replace(r', \d+:\d+\s\w{2}\s\w{3}$', '')
```

```python
quint_df['datetime'] = quint_df.datetime.str.replace(r', \d+:\d+\s\w{2}\s\w{3}$', '')
quint_df
```

```python
quint_df['datetime'] = pd.to_datetime(quint_df.datetime)
quint_df
```

```python
quint_df.to_csv('quint_df.csv', index=False)
```
