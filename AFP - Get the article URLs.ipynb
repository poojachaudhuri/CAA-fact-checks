## Step 3: Scraping 

```python
while True:
    try:
        # Do the scraping
        results = page.locator('.card a')
        count = await results.count()
        for i in range(count):
            links = results.nth(i)
            url = await links.get_attribute('href')
            print(url)
                

    # Click the 'Next' button
        async with page.expect_navigation():
            await page.locator('a.page-link.page-link-desktop[title="Go to next page"]').click()
        print("Pressing the button")
        print("---------")
    except: # Exception as e:
#         raise(e)
        # Exit the loop
        break
```

## Step 4: Put the result in a list

```python
afp_links = []
while True:
    try:
        # Do the scraping
        results = page.locator('.card a')
        count = await results.count()
        for i in range(count):
            links = results.nth(i)
            url = await links.get_attribute('href')
            afp_links.append(url)
                

    # Click the 'Next' button
        async with page.expect_navigation():
            await page.locator('a.page-link.page-link-desktop[title="Go to next page"]').click()
        print("Pressing the button")
        print("---------")
    except: # Exception as e:
#         raise(e)
        # Exit the loop
        break
```

```python
afp_links
```

```python
afp_urls = []
for each_link in afp_links:
    afp_urls.append(f"https://factcheck.afp.com/{each_link}")
afp_urls
```

## Step 5: Save the list as csv

```python
df = pd.DataFrame({'url': afp_urls})
df.head()
```
